```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 讀取數據
df = pd.read_csv('0_ddt_content.csv')
```

```python
# 讀取數據
df = pd.read_csv('1_ddt_content_tags.csv')
```

```python
# 讀取數據
df = pd.read_csv('5_ddt_log.csv')
```

```python
# 按 'GROUP' 列分組並計算出現次數
group_post_count = df.groupby('GROUP').size().reset_index(name='Post_Count')
# 按照發帖數量進行降序排序
group_post_count_sorted = group_post_count.sort_values(by='Post_Count', ascending=False)

# 顯示數據
print(group_post_count_sorted)

# 可視化發文數量
plt.figure(figsize=(10, 6))
sns.barplot(data=group_post_count_sorted.head(10), x='Post_Count', y='GROUP', palette='viridis')
plt.title('Top 10 Groups by Post Count')
plt.xlabel('Post Count')
plt.ylabel('Group')
plt.show()
```

```python
 檢查 'TYPE' 列以計算不同內容類型的數量
content_type_count = df['TYPE'].value_counts().reset_index()
content_type_count.columns = ['Content_Type', 'Count']

# 顯示內容類型計數
print(content_type_count)

# 可視化方案與博文數量
plt.figure(figsize=(8, 5))
sns.barplot(data=content_type_count, x='Content_Type', y='Count', palette='muted')
plt.title('Number of Proposals vs. Articles')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# 假設有一個 'VIEWS' 列表示每篇文章的觀看次數
# 先檢查數據結構
print(df.head())

# 計算每篇文章的觀看次數
popular_articles = df[['TITLE', 'VIEWS']].sort_values(by='VIEWS', ascending=False)

# 顯示前10篇最受歡迎的文章
print(popular_articles.head(10))

# 可視化最受歡迎的文章
plt.figure(figsize=(12, 6))
sns.barplot(data=popular_articles.head(10), x='VIEWS', y='TITLE', palette='plasma')
plt.title('Top 10 Most Popular Articles by Views')
plt.xlabel('Number of Views')
plt.ylabel('Article Title')
plt.show()
```

```python
# 假設有一個 'CAT2' 列表示第二層分類，並且有一個 'VIEWS' 列表示每篇文章的觀看次數
# 計算每個第二層分類的總觀看次數
popular_categories_cat2 = df.groupby('CAT2')['VIEWS'].sum().reset_index()

# 按觀看次數排序
popular_categories_cat2_sorted = popular_categories_cat2.sort_values(by='VIEWS', ascending=False)

# 顯示前10個熱門的第二層分類
print(popular_categories_cat2_sorted.head(10))

# 可視化最受歡迎的第二層分類
plt.figure(figsize=(12, 6))
sns.barplot(data=popular_categories_cat2_sorted.head(10), x='VIEWS', y='CAT2', palette='coolwarm')
plt.title('Top 10 Most Popular Second-Level Categories by Views')
plt.xlabel('Total Views')
plt.ylabel('Second-Level Category')
plt.show()
```

```python
# 假設有一個 'USER_TYPE' 列表示用戶類型（例如 '外部' 和 '內部'）
# 計算外部用戶與內部用戶的數量
user_type_count = df['USER_TYPE'].value_counts().reset_index()
user_type_count.columns = ['User_Type', 'Count']

# 顯示用戶類型計數
print(user_type_count)

# 計算比例
total_users = user_type_count['Count'].sum()
user_type_count['Ratio'] = user_type_count['Count'] / total_users * 100

# 可視化用戶類型比例
plt.figure(figsize=(8, 5))
plt.pie(user_type_count['Count'], labels=user_type_count['User_Type'], autopct='%1.1f%%', startangle=90)
plt.title('User Type Ratio: External vs Internal')
plt.axis('equal')  # 確保圓形
plt.show()
```
